```python

import time
from pandas_datareader import data
test=pd.read_csv('stock_industry.csv')
test=test.drop_duplicates()
ticks=test['stock'].values.tolist()
d2=pd.DataFrame()
for i in range(0,len(ticks)):
    try:
        d1=data.get_quote_yahoo(ticks[i]).reset_index()
        d2=d2.append(d1)
    except:
        print("sleeping")
        #time.sleep(300)
        print(d2.shape)
        next
        
data2=test.merge(d2,left_on='stock',right_on='index')
```

```python
data2
```

```python
ticks=test['stock'].values.tolist()
final=pd.DataFrame()
for i in range(0,len(ticks)):
    try:
        a=si.get_stats(ticks[i])
        a.set_index('Attribute', inplace=True)

        a1=pd.DataFrame(a.transpose())
        a1['stock']=ticks[i]

        final=final.append(a1)
    except:
        print("sleeping")
        #time.sleep(300)
        print(final.shape)
        next
final_df=data2.merge(final,on='stock',how='outer').fillna('')
final_df=final_df.sort_values(by=['industry','marketCap'],ascending=False)
final_df.to_csv('stocks_data.csv',index=False)
```

## Quarterly 

```python
final_data=pd.DataFrame()
for i in scrips:
    print(i)
    try:
        r = requests.get('https://financialmodelingprep.com/api/v3/income-statement/'+i+'?period=quarter&limit=5&apikey=261c92b3a1319563e8e71c6e92edd3d9')
        x = r.json()
        df = pd.DataFrame(x)
       
        income=df[['date','symbol','revenue','grossProfit','netIncome']]
        income.columns=['date','symbol','revenue_Q','grossProfit_Q','netProfit_Q']
        income['trailing_period'] = np.arange(len(income))+1
        income['trailing_period']='Trailing_Quarter_'+income['trailing_period'].astype('str')

        a=income.pivot_table(index='symbol',columns='trailing_period').reset_index()
        income_qtr=a
        
        


        r = requests.get('https://financialmodelingprep.com/api/v3/cash-flow-statement/'+i+'?period=quarter&limit=5&apikey=261c92b3a1319563e8e71c6e92edd3d9')
        x = r.json()
        df = pd.DataFrame(x)
        
        cf=df[['date','symbol','netCashProvidedByOperatingActivities','freeCashFlow']]
        cf.columns=['date','symbol','OP_Cash_Q','FCF_Q']
        cf['trailing_period'] = np.arange(len(cf))+1
        cf['trailing_period']='Trailing_Quarter_'+cf['trailing_period'].astype('str')

        a=cf.pivot_table(index='symbol',columns='trailing_period').reset_index()
        cf_qtr=a
        
        
        r = requests.get('https://financialmodelingprep.com/api/v3/income-statement/'+i+'?limit=5&apikey=261c92b3a1319563e8e71c6e92edd3d9')
        x = r.json()
        df = pd.DataFrame(x)
        income=df[['date','symbol','revenue','grossProfit','netIncome']]
        income.columns=['date','symbol','revenue_Y','grossProfit_Y','netProfit_Y']
        income['trailing_period'] = np.arange(len(cf))+1
        income['trailing_period']='Trailing_Year_'+income['trailing_period'].astype('str')
        a=income.pivot_table(index='symbol',columns='trailing_period').reset_index()
        income_ann=a


        r = requests.get('https://financialmodelingprep.com/api/v3/cash-flow-statement/'+i+'?limit=5&apikey=261c92b3a1319563e8e71c6e92edd3d9')
        x = r.json()
        df = pd.DataFrame(x)
        cf=df[['date','symbol','netCashProvidedByOperatingActivities','freeCashFlow']]
        cf.columns=['date','symbol','OP_Cash_A','FCF_A']
        cf['trailing_period'] = np.arange(len(cf))+1
        cf['trailing_period']='Trailing_Year_'+cf['trailing_period'].astype('str')

        a=cf.pivot_table(index='symbol',columns='trailing_period').reset_index()
        cf_ann=a
        test_qtr=cf_qtr.merge(income_qtr,on='symbol')
        test_ann=cf_ann.merge(income_ann,on='symbol')

        test_final=test_qtr.merge(test_ann,on='symbol')
        final_data=final_data.append(test_final)
        time.sleep(1)
    except:
        print('error')
        time.sleep(2)
```
